3.Методи класу

застосуйте методи класу до всіх обєктів a1-a3

```python
class MyAnimals:
    def __init__(self, animals, name, age, mass=None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name}, якій {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        # Правильна обробка mass, якщо воно None
        self.mass = 0 if mass is None else mass

    # Метод, що повертає звук, який видає тваринка
    def speak(self):
        if self.animals == "Кіт":
            return f"Моя Тваринка {self.animals} говорить **Мяууу**"
        elif self.animals == "Собака":
            return f"Моя Тваринка {self.animals} говорить **Гавв**"
        else:
            return f"Моя Тваринка {self.animals} не розмовляє" # для Рибок
    
    # Метод, що змінює стан об'єкта (збільшує вік)
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.name} ({self.animals}) прожила зі мною ще один рік. Тепер їй {self.age} років."

# --- Створення об'єктів ---
print("--- Створення об'єктів ---")
a1 = MyAnimals("Рибки", "Неони", 1)
a2 = MyAnimals("Кіт", "Мурка", 5, 10)
a3 = MyAnimals(animals="Собака", name="Рекс", age=2)

print("\n--- Застосування методу 'speak' до всіх об'єктів ---")
# Застосовуємо метод speak() до кожного об'єкта
print(f"**a1 (Рибки):** {a1.speak()}")
print(f"**a2 (Кіт):** {a2.speak()}")
print(f"**a3 (Собака):** {a3.speak()}")

print("\n--- Застосування методу 'add_one_year' до всіх об'єктів ---")
# Початковий вік
print(f"Початковий вік a1: {a1.age}, a2: {a2.age}, a3: {a3.age}")

# Застосовуємо метод add_one_year() до кожного об'єкта
print(a1.add_one_year())
print(a2.add_one_year())
print(a3.add_one_year())

print("\n--- Перевірка нового віку (зміна стану об'єкта) ---")
# Перевірка віку після виклику методу
print(f"Новий вік a1: **{a1.age}**")
print(f"Новий вік a2: **{a2.age}**")
print(f"Новий вік a3: **{a3.age}**")
```

Змінні класу, глобальні змінні над всіма обєктами

задайте декілька обєктів в циклі та виведіть глобальну змінну

```python
class MyAnimals:
    # --- Змінні Класу (Спільні для всіх об'єктів) ---
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age, mass=None) -> None:
        print(f"Створено: {animals} на ім'я {name}, віком {age} років")
        
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass
        
        # Оновлення змінних класу під час створення кожного об'єкта
        MyAnimals.total_animal_types.add(animals)
        MyAnimals.total_animals += 1

# --- Дані для створення об'єктів у циклі ---
animal_data = [
    ("Рибки", "Неони", 1),
    ("Кіт", "Мурка", 5),
    ("Собака", "Рекс", 2),
    ("Кіт", "Сніжок", 3), # Другий Кіт
    ("Папуга", "Кеша", 7)
]

print("--- Створення об'єктів у циклі ---")
objects_list = []
for data in animal_data:
    # Створення об'єкта та додавання його до списку
    obj = MyAnimals(data[0], data[1], data[2])
    objects_list.append(obj)

print("\n--- Виведення Глобальних Змінних Класу ---")
# Доступ до змінних через сам клас (рекомендований спосіб)
print(f"Загальна кількість створених тваринок: **{MyAnimals.total_animals}**")
print(f"Унікальні типи тваринок: **{list(MyAnimals.total_animal_types)}**")

print("\n--- Демонстрація поведінки Змінних Класу ---")

# 1. Початковий стан:
a1 = objects_list[0]
print(f"Початковий стан a1.total_animal_types: {a1.total_animal_types}") 

# 2. Спроба змінити змінну класу через екземпляр (a1.total_animal_types = ...):
a1.total_animal_types = {"Собака"} # Створюється НОВИЙ атрибут об'єкта 'a1'
                                    # Змінна класу MyAnimals.total_animal_types НЕ змінюється!

# 3. Перевірка результатів:
print(f"\nЗмінна класу, доступ через об'єкт a1: **{a1.total_animal_types}** (Це локальний атрибут a1)")
print(f"Змінна класу, доступ через Клас MyAnimals: **{MyAnimals.total_animal_types}** (Глобальна)")
```

Статичні методи та змінна self

викличте статичний метод для кожного обєкта a1-a2

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    def add_one_year(self):
        # Звичайний (екземплярний) метод: вимагає 'self' і працює з атрибутами об'єкта
        self.age += 1
        return f"Моя Тваринка {self.name} ({self.animals}) прожила за мною один рік. Тепер їй {self.age} років."

    @staticmethod
    def call_pet():
        # Статичний метод: не приймає 'self' і не працює з атрибутами об'єкта
        return "Тваринка біжить до нас"

# --- Створення об'єктів ---
a1 = MyAnimals("Рибки", "Гупії", 1)
a2 = MyAnimals("Кіт", "Мурка", 5)

print("\n--- 1. Виклик Статичного Методу (call_pet) ---")

# Виклик через об'єкт a1
print(f"Ми кличемо {a1.name} (з об'єкта): **{a1.call_pet()}**")
# Виклик через об'єкт a2
print(f"Ми кличемо {a2.name} (з об'єкта): **{a2.call_pet()}**")
# Виклик через клас (для порівняння)
print(f"Виклик з Класу MyAnimals: **{MyAnimals.call_pet()}**")


print("\n--- 2. Демонстрація 'self' та звичайних методів ---")

# Звичайний виклик методу add_one_year() через об'єкт a1
print(f"До виклику: {a1.name} вік {a1.age}")
print(f"Викликаю метод з об'єкту a1: **{a1.add_one_year()}**")
print(f"Після виклику: {a1.name} вік {a1.age}")

# Спроба викликати звичайний метод через клас, передаючи об'єкт як 'self'
# MyAnimals.add_one_year(a2) - це еквівалентно a2.add_one_year()
print(f"\nДо виклику: {a2.name} вік {a2.age}")
print(f"Викликаю метод з Класу, передаючи a2 як 'self': **{MyAnimals.add_one_year(a2)}**")
print(f"Після виклику: {a2.name} вік {a2.age}")
```

Робота з properties

додайте ще одну properties та виведіть її

```python
from datetime import datetime

class MyAnimals:
    max_age = 20
    
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    # Перша властивість: обчислює рік, до якого тварина повинна дожити
    @property
    def live_untill(self):
        # Поточний рік - поточний вік + максимальний вік
        return datetime.today().year - self.age + self.max_age
    
    # --- НОВА ВЛАСТИВІСТЬ ---
    @property
    def is_old(self):
        # Обчислює, чи вік тварини перевищує половину максимального віку
        half_max_age = self.max_age / 2
        return self.age > half_max_age
    # -----------------------
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.name} ({self.animals}) прожила за мною один рік."

# --- Створення об'єкта ---
a1 = MyAnimals("Кіт", "Мурка", 5) # Вік: 5

print(f"--- Початковий стан ({a1.name}) ---")
print(f"Поточний вік: {a1.age}")

# Виклик першої properties
print(f"Мій {a1.animals} будже зі мною до {a1.live_untill} року")
# Виклик нової properties
print(f"Чи є {a1.name} старою тваринкою (max_age/2 = 10)? **{a1.is_old}**") 

# --- Зміна стану об'єкта ---
print("\n--- Пройшло 6 років ---")
# Збільшуємо вік на 6 (стає 11 років)
for _ in range(6):
    a1.add_one_year() 
    
print(f"Поточний вік: {a1.age}")

# Повторний виклик обох properties
print(f"Мій {a1.animals} будже зі мною до {a1.live_untill} року (рік змінився)")
print(f"Чи є {a1.name} старою тваринкою тепер? **{a1.is_old}** (результат змінився)")
```

Методи Класу

допишіть метод класу який буде превіряти що атрибут name заданий з великої букви

```python
class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)

    # --- НОВИЙ МЕТОД КЛАСУ ---
    @classmethod
    def check_name_case(cls, name: str) -> str:
        """Перевіряє, чи ім'я починається з великої літери, і виправляє, якщо ні."""
        # Використовуємо .capitalize(), щоб перетворити перший символ на велику літеру, а решту на малі
        if not name[0].isupper():
            print(f"Попередження: Ім'я '{name}' було виправлено на '{name.capitalize()}'")
            return name.capitalize()
        return name
    # --------------------------

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        # Застосовуємо метод класу до вхідного імені
        self.name = MyAnimals.check_name_case(name) 
        self.age = age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data:str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        # Тут також використовується __init__, тому перевірка регістру відбудеться автоматично
        return cls(animal.strip(), name.strip(), int(age.strip()))

# --- Тестування (Створення об'єктів з некоректним регістром) ---

print("--- Створення об'єктів ---")
# Створення з CSV: ім'я "мурКа"
a1 = MyAnimals.from_csv("Кіт,мурКа,5") 

# Створення напряму: ім'я "реКс"
a2 = MyAnimals("Собака", "реКс", 2)

# Створення напряму: ім'я "Неони" (правильний регістр)
a3 = MyAnimals("Рибки", "Неони", 1)


print("\n--- Результати ---")
print(f"Мій {a1.animals}, звати {a1.name}, має {a1.age} років")
print(f"Мій {a2.animals}, звати {a2.name}, має {a2.age} років")
print(f"Мій {a3.animals}, звати {a3.name}, має {a3.age} років")
print(f"\nЗагальна кількість обєктів {MyAnimals.total_objects} а створених з csv формату {MyAnimals.total_csv}")
```

Магічні методи

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __repr__(self) -> str:
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    def __str__(self) -> str:
        return f"Це є обєкт {self.__repr__()}"

a1 = MyAnimals("Собака", "Рекс", 2)
print(a1)

print(repr(a1))

str(a1)
```
